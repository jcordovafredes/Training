```python
##Para dos o mas lugares
geo = Nominatim(user_agent="MyApp",timeout=2)
lugar_list = ("New York", "Londres", "Concepci√≥n", "Bilbao", "Pekin")
```

```python
##Como crear un mapa en m√°s de una ubicaci√≥n
plt.figure(figsize=(20,16))
my_map=Basemap(projection="ortho",lon_0=0, lat_0=0) ##Robin, ortho, etc
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color="gray")
my_map.drawmapboundary(fill_color="aqua")

#Recorrer la lista de ciudades 1 x 1
for i in lugar_list:
    address=geo.geocode(i, timeout=2)
    time.sleep(0.5)
    x,y = my_map(address.longitude, address.latitude)
    my_map.plot(x,y,color="g", marker="o", markersize="20")
```
